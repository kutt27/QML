### Entangling capability

#### Circuit with Meyer-Wallach measure of 0 

```python
from qiskit.circuit import QuantumCircuit, Parameter
theta = Parameter('θ')

#circuit A
qc = QuantumCircuit(4)
qc.rz(theta, 0)
qc.rx(theta, 0)
qc.rz(theta, 1)
qc.rx(theta, 1)
qc.rz(theta, 2)
qc.rx(theta, 2)
qc.rz(theta, 3)
qc.rx(theta, 3)

qc.draw()
```

#### Circuit with Meyer-Wallach measure more than 0 <= 1

```python

from qiskit.circuit import QuantumCircuit, Parameter
theta = Parameter('θ')

#circuit B
qc = QuantumCircuit(4)
qc.rz(theta, 0)
qc.rx(theta, 0)
qc.rz(theta, 1)
qc.rx(theta, 1)
qc.rz(theta, 2)
qc.rx(theta, 2)
qc.rz(theta, 3)
qc.rx(theta, 3)
qc.cx(0,1)
qc.cx(1,2)
qc.cx(2,3)

qc.draw()
```
